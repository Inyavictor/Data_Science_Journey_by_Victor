EXERCISE:LEVEL 3
Write an example for different Python data types such as Number(Interger, Float, Complex), Strings, Boolean, List, Tuple, set and Dictionary.


```python
#Number(Integer, Float, Complex)
Age =29
Percentage = 99.9
Algebra = 2 + 3j
print(type(Age))
print(type(Percentage))
print(type(Algebra))
```

```python
#String
FirstName = 'Victor'
print(type(FirstName))
```

```python
#Boolean
Probability = 5>4
print(type(Probability))
```

```python
#list
Programming_Languages = ['Python', 'JavaScript', 'C++', 'Java', 'Go']
print(type(Programming_Languages))
```

```python
#Tuple
Days = ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')
print(type(Days))
```

```python
#Set
Unique_Languages = {'Python', 'JavaScript', 'C++', 'Java', 'Go'}
print(type(Unique_Languages))
```

```python
#Dictionary
Me = {
    'first_name': 'Victor',
    'last_name': 'Inya',
    'age': 29,
    'country': 'Nigeria',
    'is_married': False,
    'skills': ['Python', 'JavaScript', 'C++', 'Java', 'Go'],
    'address': {
        'street': '123 Main St',
        'city': 'Lagos',
        'postal_code': '100001'
    }
}
print(type(Me))
```

```python
#EUCLIDIAN DISTANCE
import math
point1 = (2, 3)
point2 = (10,8)
distance = math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)
print(distance)
```
